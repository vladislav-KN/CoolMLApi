```python
import numpy as np
import pandas as pd
# Задаем количество дней и работников
num_days = 5
num_workers = 746  # Задайте желаемое количество работников

# Создаем массив идентификаторов работников
worker_ids = np.arange(num_workers)

# Создаем массив случайных чисел для ролей (задайте диапазон по вашему усмотрению)
roles = np.random.randint(1, 6, size=(num_workers))

# Создаем набор данных
daily_data = [{"id": worker_id, "role": role} for worker_id, role in zip(worker_ids, roles[:])]
 
# Выводим результат
dataframes_for_days = [pd.DataFrame(daily_data, columns=["id", "role"]) for i in range(num_days)]
 
```

```python
def generate_numver_of_all_msg(row):
    match row['role']:
        case 1:
            return np.random.randint(100,200)
        case 2:
            return np.random.randint(70,150)
        case 3:
            return np.random.randint(40,90)
        case 4:
            return np.random.randint(30,70)
        case 5:
            return np.random.randint(50,120)
for df in dataframes_for_days:
    df['allmsg'] = df.apply(generate_numver_of_all_msg, axis=1)
    df['allrecivedmsg'] = np.random.randint(10, 200, size=(num_workers))
```

```python
def rand_days(max):
    if(max == 0):
        return 0
    if np.random.rand()>0.9:
        return np.random.randint(max//2, max)
    elif np.random.rand()>0.3:
        return np.random.randint(0, max//3 + 1)
    else:
        return np.random.randint(max//3, max//2 + 1)
        

def generate_numver_of_day_msg(row):
    if 'day4' in df.columns:
        return row['allmsg'] - (row['day4'] + row['day3'] + row['day2'] + row['day1'])
    elif 'day3' in df.columns:
        max = row['allmsg'] - (row['day3'] + row['day2'] + row['day1'])
        return rand_days(max)
    elif 'day2' in df.columns:
        max = row['allmsg'] - (row['day2'] + row['day1']) 
        return rand_days(max)
    elif 'day1' in df.columns:
        max = row['allmsg'] - row['day1'] 
        return rand_days(max)
    else:
        return rand_days(row['allmsg'])
         
df = dataframes_for_days[0]
df['day1'] = df.apply(generate_numver_of_day_msg, axis=1)
df['day2'] = df.apply(generate_numver_of_day_msg, axis=1)
df['day3'] = df.apply(generate_numver_of_day_msg, axis=1)
df['day4'] = df.apply(generate_numver_of_day_msg, axis=1)
df['day5'] = df.apply(generate_numver_of_day_msg, axis=1)
df
```

```python
def generate_number_of_day_msg_receaved(row):

    if 'day41' in df.columns:
        return row['allrecivedmsg'] - (row['day41'] + row['day31'] + row['day21'] + row['day11'])
    elif 'day31' in df.columns:
        max = row['allrecivedmsg'] - (row['day31'] + row['day21'] + row['day11'])
        return rand_days(max)
    elif 'day21' in df.columns:
        max = row['allrecivedmsg'] - (row['day21'] + row['day11']) 
        return rand_days(max)
    elif 'day11' in df.columns:
        max = row['allrecivedmsg'] - row['day11'] 
        return rand_days(max)
    else:
        return rand_days(row['allrecivedmsg'])
df['day11'] = df.apply(generate_number_of_day_msg_receaved, axis=1)
df['day21'] = df.apply(generate_number_of_day_msg_receaved, axis=1)
df['day31'] = df.apply(generate_number_of_day_msg_receaved, axis=1)
df['day41'] = df.apply(generate_number_of_day_msg_receaved, axis=1)
df['day51'] = df.apply(generate_number_of_day_msg_receaved, axis=1)
df
```

```python
def randomize(msg_num, min, max):
    if(msg_num>40):
        return np.random.randint(min*msg_num//40, max*msg_num//40)
    elif(msg_num<5):
        return np.random.randint(min*msg_num//5, max*(1+msg_num//5))
    else:
        return np.random.randint(min, max)

def number_of_adr(row):
    msg_num = row[names[0]]
    if msg_num == 0:
        return 0
    match row['role']:
        case 1:
            
            return randomize(msg_num,5,15)
        case 2:
            return randomize(msg_num,3,10)
        case 3:
            return randomize(msg_num,2,8)
        case 4:
            return randomize(msg_num,1,5)
        case 5:
            return randomize(msg_num,5,15)
    
 

index = inc(i)
for df in new_dataframes:
    df[names[index]] = df.apply(number_of_adr, axis=1)
df
```

```python

# Масштабируйте данные (стандартизация)
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

df_cleaned = res.replace([np.inf, -np.inf], np.nan).dropna()
df_cleaned.apply(lambda col: col.drop_duplicates().reset_index(drop=True))
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_cleaned)

 
num_clusters = 2

names = [
    'Количество отправленных сообщений за период',
    'Соотношение Полученных/Отправленных',
    'Количество адресатов в отправляемых сообщениях',
    'Количество сообщений с адресатами в поле «скрытая копия»',
    'Количество сообщений с адресатами в поле «копия»',
    'Количество сообщений, прочитанных позднее времени получения',
    'Среднее время ответа на сообщение.',
    'Количество сообщений, на которые произведен ответ',
    'Количество дней между датой получения и датой прочтения сообщения',
    'Количество сообщений, отправленных вне рамок рабочего дня',
    'Количество сообщений, отправленных без подписи'
]
kmeans = KMeans(n_clusters=num_clusters)
df_cleaned['Cluster'] = kmeans.fit_predict( pd.DataFrame(scaled_data))


fig, axs = plt.subplots(11, 11,figsize=(100, 100))
for i in range(11):
    for j in range(11):
        #axs[i, j].scatter(scaled_data[:, i], scaled_data[:, j], c=df_cleaned['Cluster'], cmap='viridis', marker='o')
        axs[i, j].scatter(df_cleaned[names[ i]], df_cleaned[names[ j]], c=df_cleaned['Cluster'], cmap='viridis', marker='o')
        axs[i, j].scatter(kmeans.cluster_centers_[:, i], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Centroids')
        axs[i, j].set_xlabel(names[i])
        axs[i, j].set_ylabel(names[j])
plt.show()
```
